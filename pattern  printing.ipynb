# palindrome

```python
N = input()
N = N.lower()
newstring = ""
for i in range(len(N)):
    if N[i]==" ":
        continue
    if N[i] == "'":
        continue
    else:
        newstring+=N[i]

revstring = ""
for i in range(len(newstring)-1,-1,-1):
    revstring+=newstring[i]
if(revstring == newstring):
    print("True")
else:
    print("False")
```

# sandglass

```python
N = int(input())
N=N
for i in range(N):
    space=" "*i
    starnum="* "*(N-i)
    print(space+starnum)
for i in range(2,N+1):
    space=" "*(N-i)
    starnum="* "*i
    print(space+starnum)

```

# subsequence

```python
fullname = input()
subname = input()
subnamelen = len(subname)

index = 0
for i in fullname:
    if i == subname[index]:
        index+=1
    if index == len(subname):
        break
if index == subnamelen:
    print("Yes")
else:
    print("No")
        
```

# reverse the sentence

```python
S = input()

Sstr = list(map(str,S.split(" ")))
revstr = ""

for i in range(len(Sstr) -1,-1,-1):
    revstr +=Sstr[i]+" "
print(revstr)
```
